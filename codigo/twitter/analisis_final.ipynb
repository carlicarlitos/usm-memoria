# Evaluación de resultados y análisis

Luego de procesar el texto, se presentan los datos obtenidos y se realizan varios analisis

```python
import pymongo
from bson.objectid import ObjectId
import os
import random
from twython import Twython
import pandas as pd
import time

from math import log10

from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import Span
import matplotlib.pyplot as plt

output_notebook()
```

```python
#conexion mongo
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["twitter-memoria"]
coleccion_completa = mydb["nuevos_bi_tweets"]
bd_hilos = mydb["nuevos_bi_hilos"]
usuarios = mydb["nuevos_usuarios"]
topicos = mydb["nuevos_topicos"]

#obtener llaves Twitter
llaves_bd = mydb["llaves"]
llaves = llaves_bd.find_one()

APP_KEY = llaves["API_KEY"]
ACCESS_TOKEN  = llaves["ACCESS_TOKEN_T"]
twitter       = Twython(APP_KEY, access_token=ACCESS_TOKEN)
```

```python
all_tweets = [tweet for tweet in coleccion_completa.find({})]
all_hilos = [hilo for hilo in bd_hilos.find({})]
all_users = [usuario for usuario in usuarios.find({})]
all_topicos = [topico for topico in topicos.find({})]
```

```python
dftweet = pd.DataFrame(all_tweets).drop(["_id"], axis=1)
dfhilo = pd.DataFrame(all_hilos).drop(["_id"], axis=1)
dfusuario = pd.DataFrame(all_users).drop(["_id"], axis=1)
dftopico = pd.DataFrame(all_topicos).drop(["_id"], axis=1)
```

```python
no_max_tweets = dftweet
for i in range(0,4,1):
    no_max_tweets = no_max_tweets.drop(no_max_tweets['total_rt'].idxmax())
    no_max_tweets = no_max_tweets.drop(no_max_tweets['total_fav'].idxmax())
    
tweet_rt = no_max_tweets["total_rt"].tolist()
tweet_fv = no_max_tweets["total_fav"].tolist()
```

```python
grafico1 = figure(plot_width=700, 
           plot_height=350, 
           title='relacion favoritos vs retweets', 
           background_fill_color="#f2f3f7", 
           y_axis_label='likes', 
           x_axis_label='retweets')

grafico1.circle(tweet_rt, tweet_fv, line_color='#cd2626')

show(grafico1)
```

```python
log_tweet_rt = [log10(valor+1) for valor in tweet_rt]
log_tweet_fv = [log10(valor+1) for valor in tweet_fv]
```

```python
grafico1 = figure(plot_width=700, 
           plot_height=350, 
           title='relacion favoritos vs retweets', 
           background_fill_color="#f2f3f7", 
           y_axis_label='likes', 
           x_axis_label='retweets')

grafico1.circle(log_tweet_rt, log_tweet_fv, line_color='#cd2626')

show(grafico1)
```
